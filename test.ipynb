```python
import random

def choose_samples(anchor, items):
    anchor_orientation = anchor["orientation"]
    anchor_id = anchor["p_id"]
    same_id, diff_id = [], []
    for item in items:
        if item['p_id'] == anchor_id:
            same_id.append(item)
        else:
            diff_id.append(item)

    same_id_diff_ori, diff_id_same_ori = [], []

    if 0 <= anchor_orientation <= 9 or 63 <= anchor_orientation <= 71 or 27 <= anchor_orientation <= 45:
        # anchor has back or front orientation
        for item in same_id:
            if 45 <= item["orientation"] < 63 or 9 <= item["orientation"] < 27:
                    # found positive sample of sideway orientation
                same_id_diff_ori.append(item)
        for item in diff_id:
            if 0 <= item["orientation"] <= 9 or 63 <= item["orientation"] <= 71 or 27 <= item["orientation"] <= 45:
                diff_id_same_ori.append(item)
    else:
        # anchor has sideway orientation
        for item in same_id:
            if 0 <= item["orientation"] <= 9 or 63 <= item["orientation"] <= 71 or 27 <= item["orientation"] <= 45:
                    # found positive sample of back orientation
                same_id_diff_ori.append(item)
        for item in diff_id:
            if 45 <= item["orientation"] < 63 or 9 <= item["orientation"] < 27:
                diff_id_same_ori.append(item)
    positive_sample = random.choice(same_id_diff_ori)
    negative_sample = random.choice(diff_id_same_ori)
    return positive_sample, negative_sample
```

```python

import json 

with open('/home/dustin/Documents/Research/P003 - 2D ReID/Code/person-reid/data/market1501/jsons/train.json', 'rb') as f:
    img_list = json.load(f)
```

```python
anchor_item = {
    "p_id": 2,
    "img_path": "0002_c2s1_136741_01.jpg",
    "orientation": 3,
}


positive_sample, negative_sample = choose_samples(anchor_item, img_list)

print("Positive sample:", positive_sample)
print("Negative sample:", negative_sample)
```
